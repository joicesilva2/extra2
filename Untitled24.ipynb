```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
data = {
    'País': ['Brasil', 'Estados Unidos', 'China', 'Japão', 'Alemanha'],
    'População': [211, 328, 1393, 126, 83],
    'PIB': [21477, 21433, 14343, 5080, 4217],
    'Área': [8515767, 9629091, 9640011, 377972, 357582]
}

# Criando um DataFrame com os dados
df = pd.DataFrame(data)

# Gráfico de barras simples
plt.figure(figsize=(8, 6))
plt.bar(df['País'], df['População'])
plt.xlabel('País')
plt.ylabel('População')
plt.title('População por País')
plt.show()

# Gráfico de barras agrupados
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = df.index
plt.bar(index, df['População'], bar_width, label='População')
plt.bar(index + bar_width, df['PIB'], bar_width, label='PIB')
plt.xlabel('País')
plt.ylabel('Valor')
plt.title('População e PIB por País')
plt.xticks(index + bar_width / 2, df['País'])
plt.legend()
plt.show()

# Gráfico de barras empilhadas
plt.figure(figsize=(8, 6))
plt.bar(df['País'], df['População'], label='População')
plt.bar(df['País'], df['PIB'], bottom=df['População'], label='PIB')
plt.xlabel('País')
plt.ylabel('Valor')
plt.title('População e PIB por País')
plt.legend()
plt.show()

# Gráfico de bolhas
plt.figure(figsize=(8, 6))
plt.scatter(df['Área'], df['PIB'], s=df['População']*10, alpha=0.5)
plt.xlabel('Área')
plt.ylabel('PIB')
plt.title('Área vs. PIB (Tamanho das bolhas representa a População)')
plt.show()

# Gráfico de linhas
plt.figure(figsize=(8, 6))
plt.plot(df['País'], df['População'], marker='o', label='População')
plt.plot(df['País'], df['PIB'], marker='s', label='PIB')
plt.xlabel('País')
plt.ylabel('Valor')
plt.title('População e PIB por País')
plt.legend()
plt.show()

# Gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(df['Área'], df['PIB'], s=df['População']*10, alpha=0.5)
plt.xlabel('Área')
plt.ylabel('PIB')
plt.title('Área vs. PIB (Tamanho das bolhas representa a População)')
plt.show()

# Dados de exemplo para a generosidade dos países
data_generosidade = {
    'Country': ['Brazil', 'United States', 'China', 'India', 'Japan'],
    'Generosidade': [0.6, 0.4, 0.3, 0.2, 0.5]
}

# Criar um DataFrame com os dados de generosidade
df_generosidade = pd.DataFrame(data_generosidade)

# Gráfico de generosidade
plt.figure(figsize=(8, 6))
plt
```
